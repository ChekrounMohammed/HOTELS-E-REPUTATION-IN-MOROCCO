# recleaning tweets and sentiment analysis

```python
tweets=pd.read_csv('Tweets_Rev_Final.csv',encoding='utf-8',sep=';')
```

```python
tweets=tweets.drop_duplicates()
```

```python
tweets=tweets.dropna()
```

```python
tweets
```

```python
from textblob import TextBlob 
```

```python
res = TextBlob('the movie is goodly described')
print(res.sentiment.polarity)
```

```python
def analyseSentiment(sentence):
    res=TextBlob(str(sentence))
    an=res.sentiment.polarity
    if(an<0):
        return('neg')
    elif an==0:
        return('ntr')
    else:
        return('pos')
```

```python
tweets['classification']=tweets.apply(lambda row : analyseSentiment(row['review']), axis = 1)
```

```python
def cleanrev(rev):
    rev=str(rev)
    rev.replace('@','')
    rev.replace('#','')
    rev.replace('RT','')
    rev.replace('\n','')
    rev.replace('"',"")
    rev.replace('(','')
    rev.replace(')','')
    rev.replace("'",'')
    rev.replace("\\",'')
    rev.replace('*','')
    rev.replace(';',' ')
    rev.replace(',',' ')
tweets.apply(lambda row : cleanrev(row['review']), axis = 1)
```
